```python
indir = "/media/manikya_varshney/My Passport/Mani(Yale)/Data/h01-20200818-20200824/h01-20200819"
outfile = "/media/manikya_varshney/My Passport/Mani(Yale)/Data/h01-20200818-20200824/August19th.csv"
```

```python
def concatenate(indir, outfile):
    os.chdir(indir)
    fileList = glob.glob("*.csv")
    dfList = []
    for filename in fileList:
        print(filename)
        df = pd.read_csv(filename, header = 0, low_memory=False)
        dfList.append(df)
    print(len(dfList))
    concatDf = pd.concat(dfList, axis = 0)
    #concatDf.to_csv(outfile, index = None)
    concatDf.to_parquet(outfile, compression='gzip')  
```

```python
concatenate(indir, outfile)
```
